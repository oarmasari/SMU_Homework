```python
import pandas as pd
```

```python
my_csv_dfs = []
for n in range (1,12):
    url = f"{n}-citibike.csv"
    my_csv_dfs.append(pd.read_csv(url))
```

```python
main_df=pd.concat(my_csv_dfs)
```
