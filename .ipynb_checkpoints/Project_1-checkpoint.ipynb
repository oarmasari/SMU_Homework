```python
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import requests
import io
```

```python
# File to Load (Remember to Change These)
file_to_load = "yelp_data.csv"
```

```python
# Read Purchasing File and store into Pandas data frame
yelp_data = pd.read_csv(file_to_load, low_memory=False)
yelp_data.head(15)
```

```python
# Filter the DataFrame down only to those columns to chart
review_count = yelp_data[["state","review_count"]]

# Set the index to be "State" so they will be used as labels
#review_count = review_count.set_index("state")

review_count.head()
```

```python
reviewGroups = review_count.groupby("state").sum().reset_index()
reviewGroups.head()
```

```python
len(reviewGroups)
```

```python
reviewGroupsSort = reviewGroups.sort_values("review_count", ascending=False).reset_index(drop=True)
reviewGroupsSort.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
reviewGroupsSort.plot(kind="bar", x="state", y="review_count", figsize=(10,6))

# Set a title for the chart
plt.title("Reviews by State")

plt.show()
plt.tight_layout()
```

```python
plt.close()
```

```python
%matplotlib inline

plt.figure(figsize=(20,10))

plt.bar(reviewGroupsSort.state, np.log(reviewGroupsSort.review_count), color="purple")


plt.title("Review Counts by State", weight="bold", fontsize=20)
plt.xlabel("States", fontsize=16)
plt.ylabel("Number of Reviews (Log)", fontsize=16)

plt.grid(axis='y')

plt.savefig("reviewsByState.png")


plt.show()
```
